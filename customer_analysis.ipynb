```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Total sales by Prouduct line
df = pd.read_csv("superMarket.csv")
plt.figure()
sns.barplot(x='Product line', y='gross income', data=df, estimator=sum)
plt.xticks(rotation=45)
plt.title("Total Sales by Product Line")
plt.ylabel("Gross Income")
plt.show()
```

```python
plt.figure()
sns.barplot(x='Customer type',y='gross income',data=df,estimator=sum,palette=['blue', 'pink'] )
plt.title("Sales by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Total Gross Income")
plt.show()
```

```python
#Trend line
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
monthly_sales = df.groupby(df['Date'].dt.to_period("M"))['gross income'].sum()
plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Gross Income")
plt.xticks(rotation=45)
plt.show()
```

```python
#spending segmentation
df['Spending_Level'] = pd.qcut(df['gross income'], 3, labels=['Low', 'Medium', 'High'])
plt.figure()
sns.countplot(x='Spending_Level', data=df)
plt.title("Customer Spending Segmentation")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Gender', y='gross income', data=df)
plt.title("Spending by Gender")
plt.show()
```
